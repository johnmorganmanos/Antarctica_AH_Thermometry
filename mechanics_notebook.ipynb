### Can we do a rolling correlation?

```python
cutoff = 80 #Cut out the seasonal signal
pf = PolynomialFeatures(degree=6)
md2 = LinearRegression()  
sos = signal.butter(10, 1/50, 'lp', fs=0.255, output='sos')

window_size = 10
half_size = int(window_size/2) #must be an integer value

fig,axes = plt.subplots(1,len(data_proc), figsize=[15,5])

for num, (i, ax) in enumerate(zip(data_proc, axes)):
    x=np.asarray(data_proc[i][0]-data_proc[i][0][0])
    x=x.reshape(len(x),1)
    x = x[cutoff:]
    Xp = pf.fit_transform(x)
    md2.fit(Xp, anomalies[i][cutoff:])
    trendp = md2.predict(Xp)
#     ax.plot(anomalies[i][cutoff:],x)
#     ax.plot(trendp,x)
    alpha = detrend(data_proc[i][-1][cutoff:])
    anomalies_detrend = anomalies[i][cutoff:]-trendp
    
    anomalies_filtered = signal.sosfilt(sos, anomalies_detrend)
    alpha_filtered = signal.sosfilt(sos, alpha)

    correlation = np.zeros(len(x))
    for k in range(len(x)-window_size):
        
        correlation[k+half_size] = pearsonr(anomalies_filtered[k:k+window_size], 
                                            alpha_filtered[k:k+window_size])[0]
        

    ax.plot(correlation, x*-1)
    ax1 = ax.twiny()
    ax1.plot(anomalies_filtered,x*-1, c='orange')
#     ax1 = ax.twiny()
#     ax1.plot(alpha_filtered[cutoff:], x[cutoff:]*-1, c='orange')
    ax.set_ylim([-100,-90])
    ax.set_title(i[0:12])
    if num != 0:
        ax.set_yticklabels([])
    if num == 0:
        ax.set_ylabel('Depth [m]')
        ax.set_xlabel('Correlation coeffecient')
```
