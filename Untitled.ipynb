# Import packages

```python
import numpy as np
from heat import *
import matplotlib.pyplot as plt
from scipy.signal import detrend
import pickle as pkl
```

# Import real data

```python
filepath = '/data/fast1/jm-scratch/data_proc.pkl'
file = open(filepath, 'rb')
data = pkl.load(file)
```

# Model parameters

```python
tmax = 2023
tmin = 1923
zmax = 200
dTdz = 0.02
nz = 99
nt = 99
alpha = 35
```

# Solve, no ablation

```python

accumulation = 0

t_surf = np.array([-31]*nt)
U,t,z = heat(t_surf,
         tmax = tmax,
         tmin = tmin,
         zmax = zmax,
         dTdz = dTdz,
         nz = nz,
         nt = nt,
         alpha = alpha,
         accumulation = accumulation)
```

```python
fig, ax = plt.subplots()
im = ax.pcolormesh(t,z,U)
fig.colorbar(im,ax=ax)
plt.gca().invert_yaxis()

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(U[:,-1],z*-1)
ax2.plot(detrend(U[:,-1]),z*-1)
```

# Solve,with a constant ablation

```python

accumulation = -0.1

t_surf = np.array([-31]*nt)
U,t,z = heat(t_surf,
         tmax = tmax,
         tmin = tmin,
         zmax = zmax,
         dTdz = dTdz,
         nz = nz,
         nt = nt,
         alpha = alpha,
         accumulation = accumulation)
```

# Solve, with variable ablation

```python
accumulation = np.linspace(0.1,-0.1,nt+1)
t=np.linspace(tmin,tmax, nt+1)

fig,ax = plt.subplots()
ax.plot(t,accumulation)
```

```python
t=np.linspace(tmin,tmax, nt+1)
amp = -0.2
accumulation =  amp*np.sin(2*np.pi*0.001*t)

fig,ax = plt.subplots()
ax.plot(t,accumulation)
ax.set_ylabel('Accumulation')
```

```python
t_surf = np.array([-31]*nt)
U,t,z = heat_var_accum(t_surf,
                       accumulation,
         tmax = tmax,
         tmin = tmin,
         zmax = zmax,
         dTdz = dTdz,
         nz = nz,
         nt = nt,
         alpha = alpha)
```

```python
fig, ax = plt.subplots()
im = ax.pcolormesh(t,z,U)
fig.colorbar(im,ax=ax)
plt.gca().invert_yaxis()

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(U[:,-1],z*-1)
ax1.set_title('Present Temp Profile')

ax2.plot(detrend(U[:,-1]),z*-1)
ax2.set_title('Present Detrended Temp Profile')
```
