# Can we get an inverse problem?

```python
accumulation = np.eye(nt+1)
t_surf = np.array([-31]*nt)
G=np.zeros((nz+1,nt+1))
for k,i in enumerate(accumulation):
    U,t,z = heat_var_accum(t_surf,
                       i,
         tmax = tmax,
         tmin = tmin,
         zmax = zmax,
         dTdz = dTdz,
         nz = nz,
         nt = nt,
         alpha = alpha)
    G[:,k]=U[:,-1]
```

```python
fig, ax = plt.subplots()
im = ax.pcolormesh(t,z*-1,detrend(G,axis=1))
fig.colorbar(im, ax=ax)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
for i in data:
    anomaly = data[i][8].values
    inversed_realData = inverser_SVD(detrend(G, axis=1),anomaly,nz=nz+1,p=3)
    ax1.plot(t,inversed_realData[0])
    
    forward_model = inversed_realData[0] @ detrend(G, axis=1)
    ax2.plot(forward_model,z*-1,ls='--')
    ax2.plot(anomaly,data[i][0].values*-1)
ax2.set_xlabel('Temperature (K)')
        
```

# Tuning advection

### Test 1: Constant advection in time

```python
t_surf = np.array([-32]*nt)
advection_tests = np.linspace(-0.5,0.5,11)
```

```python
fig,ax = plt.subplots()

for i in advection_tests:
    accumulation = np.array([i]*nt)
    U,t,z = heat_var_accum(t_surf,
                           accumulation,
             tmax = tmax,
             tmin = tmin,
             zmax = zmax,
             dTdz = dTdz,
             nz = nz,
             nt = nt,
             alpha = alpha)
    ax.plot(U[:,-1],z*-1, label=str(i))
    
plt.legend()
```

### Test 2: Delta change at some time in the past 

```python
t_surf = np.array([-32]*nt)

advection_tests = np.arange(0,nt+1,10)
```

```python
fig,ax = plt.subplots()

for i in advection_tests:
    advection_cons = np.array([0.]*nt)
    advection_cons[i] = .5
    accumulation = advection_cons
    U,t,z = heat_var_accum(t_surf,
                           accumulation,
             tmax = tmax,
             tmin = tmin,
             zmax = zmax,
             dTdz = dTdz,
             nz = nz,
             nt = nt,
             alpha = alpha)
    ax.plot(detrend(U[:,-1]),z*-1, label='delta: '+str(100-i))
    
plt.legend()
```

### Test 3: Step change at some time in the past 

```python
t_surf = np.array([-32]*nt)

advection_tests = np.arange(0,nt+1,10)
```

```python
fig,ax = plt.subplots()

for i in advection_tests:
    advection_cons = np.array([0.]*nt)
    advection_cons[i:] = .5
    accumulation = advection_cons

    U,t,z = heat_var_accum(t_surf,
                           accumulation,
             tmax = tmax,
             tmin = tmin,
             zmax = zmax,
             dTdz = dTdz,
             nz = nz,
             nt = nt,
             alpha = alpha)
    ax.plot(detrend(U[:,-1]),z*-1, label='start: '+str(i))
    
plt.legend()
```

### Test 4: linear trend in time with varying degrees of rate of change

```python
t_surf = np.array([-32]*nt)
advection_tests = np.logspace(-3,0,11)
# fig,ax = plt.subplots()
# for i in advection_tests:
#     accumulation = np.linspace(-i,i,nt)+i
#     ax.plot(accumulation)
```

```python
fig,ax = plt.subplots()

for i in advection_tests:
    accumulation = np.linspace(-i,i,nt)+i
    U,t,z = heat_var_accum(t_surf,
                           accumulation,
             tmax = tmax,
             tmin = tmin,
             zmax = zmax,
             dTdz = dTdz,
             nz = nz,
             nt = nt,
             alpha = alpha)
    ax.plot(U[:,-1],z*-1, label=str(round(i,3)))
    
plt.legend()
```

### Test 5: sinusoidal trend with varying frequencies

```python
t=np.linspace(0,nt, nt)
amp = 0.2
accumulation =  amp*np.sin(2*np.pi*1*t)+amp

fig,ax = plt.subplots()
ax.plot(t,accumulation)
ax.set_ylabel('Accumulation')
```

```python
t_surf = np.array([-32]*nt)
advection_tests = np.logspace(-3,2,11)
fig,ax = plt.subplots()

for i in advection_tests:
    time=np.linspace(0,nt, nt)
    amp = 0.2
    accumulation =  amp*np.sin(2*np.pi*i*time)+amp
    ax.plot(accumulation)
```

```python
fig,ax = plt.subplots()

for i in advection_tests:
    time=np.linspace(0,nt, nt)
    amp = 0.2
    accumulation =  amp*np.sin(2*np.pi*i*time)+amp
    U,t,z = heat_var_accum(t_surf,
                           accumulation,
             tmax = tmax,
             tmin = tmin,
             zmax = zmax,
             dTdz = dTdz,
             nz = nz,
             nt = nt,
             alpha = alpha)
    ax.plot(detrend(U[:,-1]),z*-1, label=str(round(i,3)))
    
plt.legend()
plt.title("Frequency (per year)")
```
