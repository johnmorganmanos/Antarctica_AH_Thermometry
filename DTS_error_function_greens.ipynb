# 3) Create the Green's Function using the error function

```python
tmin = 1023
tmax = 2023
nz = 20
num_steps = 4000
```

```python
greens = {}
greens_beltrami = {}
for i in data_proc:

    borehole_proc = data_proc[i]
    green, green_bel = greens_generator_efc(borehole_proc, 
                                            tmin = tmin, 
                                            tmax = tmax, 
                                            nz = nz, 
                                            num_steps=num_steps)
    
    greens[i] = green
    greens_beltrami[i] = green_bel
```

```python

fig, axes = plt.subplots(len(greens), 1)
v=0.1
for i,ax in zip(greens, axes):
    ax.pcolormesh(greens[i], vmin=-v, vmax=v, cmap='seismic')
```

```python

fig, axes = plt.subplots(len(greens), 1)
v=0.1
for i,ax in zip(greens_beltrami, axes):
    ax.pcolormesh(greens_beltrami[i], vmin=-v, vmax=v, cmap='seismic')
```

# 4) Now do the inversion - this the SVD

```python
inverse_results = {}
borehole_temps = {}
depths = {}
cutoff = 0
for i in greens:
    z = data_proc[i][0].values[cutoff:]
    borehole_temp = data_proc[i][8].values
    borehole_temp = borehole_temp[cutoff:]
    inverse_results[i] = inverser_SVD(greens[i],
                                      borehole_temp,
                                      nz=nz,
                                      p=8)
    borehole_temps[i] = borehole_temp
    depths[i] = z
```

```python
fig, axes = plt.subplots(len(inverse_results), 3, figsize=[15,15])
t = np.linspace(tmax,tmin,num_steps-1)
for i, (ax1,ax2, ax3) in zip(inverse_results, axes):
    ax1.plot(t, inverse_results[i][0])
    ax2.plot(np.diag(inverse_results[i][2]))
    ax2.set_yscale('log')
    ax3.plot(borehole_temps[i], depths[i]*-1)
    ax1.set_xlim([1953,2023])
    ax1.set_ylim([-2,2])
```

```python
### Plot on top of each other ###
%matplotlib notebook
fig, (ax1,ax2,ax3) = plt.subplots(1, 3, figsize=[15,5])
t = np.linspace(tmax,tmin,num_steps-1)
for i in inverse_results:
    ax1.plot(t, inverse_results[i][0])
    ax2.plot(np.diag(inverse_results[i][2]))
    ax2.set_yscale('log')
    ax3.plot(borehole_temps[i], depths[i]*-1)
```

# 6) Test with synthetic data

```python
### climate signal we want to recover ###
t = np.linspace(tmin,tmax,num_steps)
syn_temp = np.zeros(len(t))
step = int(len(t)*.95)
syn_temp[step:] = .1

fig, ax = plt.subplots()
ax.plot(t, syn_temp)
ax.set_title('Synthetic temperature signal')
plt.show()
```

```python
### Foreword model ###
profiles = {}
ste_sta = {}
z_s = {}
# fig, ax = plt.subplots()
for i in data_proc:
    temp_to_model = syn_temp
#     ax.plot(temp_to_model)

    steady_state, modeled_profile, z = foreword_modeler(data_proc[i],
                                                        timeseries_toModel=temp_to_model,
                                                        tmin = tmin, 
                                                        tmax = tmax, 
                                                        num_steps = num_steps, 
                                                        nz = nz,
                                                        model_timeseries=True)
    
    profiles[i] = modeled_profile
    ste_sta[i] = steady_state
    z_s[i] = z
    
    
```

```python
fig, ax = plt.subplots()
for i in profiles:
    ax.plot(profiles[i],z_s[i]*-1)
```
