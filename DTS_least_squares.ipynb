# 4) Performing the SVD

```python
inverse_1000_all = {}

for i in greensFunctions_1000:
    boreholeTemp = steady_states_1000[i]
    greens_function = greensFunctions_1000[i]
    inverse = inverser_SVD(greens_function, boreholeTemp, p = greens_function.shape[0])
    inverse_1000_all[i] = inverse
```

```python
inverse_1_all = {}

for i in greensFunctions_1:
    boreholeTemp = steady_states_1[i]
    greens_function = greensFunctions_1[i]
    inverse = inverser_SVD(greens_function, boreholeTemp, p = greens_function.shape[0])
    inverse_1_all[i] = inverse
```

```python
### Show the singular matrix ###

fig, ax = plt.subplots()

for i in inverse_1000_all:

    ax.plot(np.diag(inverse_1000_all[i][2]), label = i+' 500 years', c='red')
    ax.set_yscale('log')
    

### Show the singular matrix ###


for i in inverse_1_all:
    
    ax.plot(np.diag(inverse_1_all[i][2]), label = i+ ' 25 years', c='blue')
    ax.set_yscale('log')

plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```
