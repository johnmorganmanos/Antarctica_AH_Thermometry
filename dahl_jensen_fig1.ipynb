# Run the heat model forward model

```python
time_steps = 500000
models = {}
for i in data_proc:
    H = data_proc[i][6]
    Tsurf = data_proc[i][7]
    models[i] = DJ_foreword(H=H, Tsurf=Tsurf, time_steps=time_steps)
```

```python
fig, axes = plt.subplots(1, len(models), figsize=([20,5]))
for k, (i,ax) in enumerate(zip(models, axes)):
    ax.plot(data_proc[i][2], data_proc[i][0]*-1, label='Data')
    ax.plot(models[i][0],models[i][1][::-1]*-1, label='Foreword Model - Iceotherm')
    if k == len(models) -1:
        ax.legend(bbox_to_anchor=[1.1,1.2])
    ax.set_title(i)
    ax.set_xlim([np.min(models[i][0]),np.max(models[i][0])])
    
```

```python
fig, axes = plt.subplots(1, len(models), figsize=([20,5]))
for k, (i,ax) in enumerate(zip(models, axes)):

    x = data_proc[i][0]
    y = np.array(data_proc[i][2].values)

    y = np.reshape(y, (y.shape[0],))
    f = interpolate.interp1d(x, y)
    xnew = models[i][1][::-1]
    ynew = f(xnew)
#     ax.plot(ynew, xnew*-1, label='Data')
#     ax.plot(models[i][0],models[i][1][::-1]*-1, label='Foreword Model - Iceotherm')
    model = models[i][0]
    residual = ynew - model
    
    ax.plot(residual, xnew*-1)
    ax.plot([0.,0.], [xnew[-1]*-1,xnew[0]*-1])
#     if k == len(models) -1:
# #         ax.legend(bbox_to_anchor=[1.1,1.2])
    ax.set_xlim([-1.5,0.5])
    ax.set_title(i)
```
